```python
import pandas as pd
```

```python
sr_df = pd.read_parquet("./cleaned_data/SubmitRecord_knowledge_ranked.parquet")
df = pd.read_parquet("./cleaned_data/TitleInfo.parquet")
```

```python
sr_df
```

```python
df
```

```python
# Count the number of passed students for each title
passed_counts = sr_df[sr_df['state'] == 'Absolutely_Correct'].groupby('title_ID').size().reset_index(name='passed_count') # type: ignore
df = pd.merge(df, passed_counts, on='title_ID', how='left')
df
```

```python
total_count = sr_df.groupby('title_ID').size().reset_index(name='total_count') # type: ignore
df = pd.merge(df, total_count, on='title_ID', how='left')
df
```

```python
avg_retries = sr_df.groupby('title_ID')['retry_count'].mean().reset_index(name='avg_retries')
df = pd.merge(df, avg_retries, on='title_ID', how='left')
df
```

```python
avg_score = sr_df.groupby('title_ID')['score'].mean().reset_index(name='avg_score')
df = pd.merge(df, avg_score, on='title_ID', how='left')
df
```

```python
avg_score_percentage = df.apply(lambda row: row['avg_score'] / row['score'], axis=1)
df['avg_score_percentage'] = avg_score_percentage
df
```

```python
df.to_parquet("cleaned_data/TitleInfo_with_stats.parquet")
```
