```python
from pathlib import Path

import pandas as pd
```

```python
files = Path("./data/Data_SubmitRecord").glob("SubmitRecord-*.csv")

df = pd.concat([pd.read_csv(f) for f in files])
# convert timestamp to pandas datetime
df["time"] = pd.to_datetime(df["time"], unit="s", origin="unix")
# drop index column
df = df.drop("index", axis=1)
df
```

```python
# list every possible values of class
df["class"].unique()
```

```python
# remove abnormal values
df = df[df["class"] != "class"]
df
```

```python
# get type of each column
df.dtypes
```

```python
# display non-int timeconsume
df[df["timeconsume"].str.isnumeric() == False]
```

```python
# give them default value
df.loc[df["timeconsume"].str.isnumeric() == False, "timeconsume"] = 0
df[df["timeconsume"].str.isnumeric() == False]
```

```python
# convert the whole timeconsume column to int
df.loc[:, "timeconsume"] = df["timeconsume"].astype(int)
```

```python
# title_ID should match Question_*, method should match Method_*
# validate them and find the abnormal values
df[(df["title_ID"].str.startswith("Question_") == False) | (df["method"].str.startswith("Method_") == False)]
```

```python
df["state"].unique()
```

```python
normal_state = [
    "Absolutely_Correct",
    "Partially_Correct",
    "Absolutely_Error",
    "Error1",
    "Error2",
    "Error3",
    "Error4",
    "Error5",
    "Error6",
    "Error7",
    "Error8",
    "Error9",
]
df = df[df["state"].isin(normal_state)]
df
```

```python
# plot memory usage
df["memory"].plot(kind="hist")
```

```python
df["memory"].describe(percentiles=[.8, .9, .95, .99, .999])
```

```python
# filter out data of memory > 10000
df[df["memory"] > 10000]
```

```python
# find if there are any other blank values
df.isnull().sum()
```

```python
# read student info
student_info = pd.read_csv("./data/Data_StudentInfo.csv")
# find any rows that don't have valid student_ID
df[~df["student_ID"].isin(student_info["student_ID"])]
```

```python
# filter out
df = df[df["student_ID"].isin(student_info["student_ID"])]
```

```python
# so do title info
title_info = pd.read_csv("./data/Data_TitleInfo.csv")
df[~df["title_ID"].isin(title_info["title_ID"])]
```

```python
# no invalid rows so good
```

```python
# output to cleaned_data folder
df.to_parquet("./cleaned_data/SubmitRecord.parquet")
student_info.drop(["index"], axis=1).to_parquet("./cleaned_data/StudentInfo.parquet")
title_info.drop(["index"], axis=1).to_parquet("./cleaned_data/TitleInfo.parquet")
```
