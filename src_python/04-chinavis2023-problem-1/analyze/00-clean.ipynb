```python
from pathlib import Path
import json

import pandas as pd
from pandarallel import pandarallel

pandarallel.initialize(progress_bar=True)
```

```python
files = Path("./data").glob("part-*.json")

df = pd.DataFrame()

for f in files:
    df_partial = pd.read_json(f, lines=True)
    df = pd.concat([df, df_partial], ignore_index=True)
    print(f"read {f}")

df
```

```python
df = df[df["seq"] > 0]
df
```

```python
cnt = df.groupby("id")["type"].nunique().reset_index()
df = df[df["id"].isin(cnt[cnt["type"] == 1]["id"])]
df
```

```python
df.loc[:,"position"] = df["position"].parallel_apply(json.loads)
df
```

```python
df.loc[:,"shape"] = df["shape"].parallel_apply(json.loads)
df
```

```python
df.loc[:,"time"] = pd.to_datetime(df["time_meas"], unit="us", origin="unix")
df = df.drop("time_meas", axis=1)
df
```

```python
df.to_parquet("./cleaned_data/road_data.parquet", index=False)
```
